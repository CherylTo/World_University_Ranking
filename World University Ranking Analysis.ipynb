There are cleaning to do before analysis can take place. <br>
Code to convert object to datatype float
#to change use .astype() 
#drinks['beer_servings'] = drinks.beer_servings.astype(float)

```python
#Total Score
times_df['total_score'] = times_df['total_score'].replace('-', '')
```

```python
times_df['total_score'] = pd.to_numeric(times_df['total_score'])
```

```python
times_df = times_df.replace('-','')
```

```python
#Income
a = times_df['income'].fillna(np.nan)
times_df['income'] = pd.to_numeric(a)
```

```python
#Num_Students
b = times_df['num_students'].str.replace(",","")
times_df['num_students'] = pd.to_numeric(b)
```

```python
#International Students
c = times_df['international'].fillna(np.nan)
times_df['international'] = pd.to_numeric(c)
```

<b> Need to write a function that will separate the first and second values of the ration and then turn them into a decimal value

```python
print(times_df['female_male_ratio'][1][0:2])
print(times_df['female_male_ratio'][1][5:7])
```

```python
f = pd.to_numeric(times_df['female_male_ratio'][1][0:2])
m = pd.to_numeric(times_df['female_male_ratio'][1][5:7])
f/m
```

```python
female = []
for i in np.arange(0,len(times_df)):
    fem = str(times_df['female_male_ratio'][i])
    fem = fem[0:2]
    fem = pd.to_numeric(fem, errors = 'coerce')
    female.append(fem)
    
```

```python
#female_ratio
female[:5]
```

```python
male = []
for i in np.arange(0,len(times_df)):
    men = str(times_df['female_male_ratio'][i])
    men = men[5:7]
    men = pd.to_numeric(men, errors = 'coerce')
    male.append(men)
```

```python
male[:5]
```

```python
m_f = pd.Series(female).divide(pd.Series(male))
m_f[:5]
```

```python
m_f
```

```python
type(m_f)
```

```python
#Adding the new ratio to the dataframe
times_df['fm_ratio'] = m_f.values
```

```python
#dropping the female_male_ratio column (we don't need it anymore)
times_df = times_df.drop('female_male_ratio', axis = 1)
```

<b> Top Universities from 2011-2016

```python
top6_times = times_df[['year', 'total_score']].groupby('year').head()
top6_s = top6_times.reset_index()
```

```python
top6_s.head()
```

```python
top6_s.pivot(index = 'year', columns = 'university_name', values = 'total_score').plot(kind = 'bar', title = 'Top 6 Scores', figsize = (15,6), rot = 'horizontal')
plt.title('Top 5 Universities from 2011 to 2016')
plt.ylabel('Year')
plt.xlabel('Total Score')
plt.ylim((85,100))
```

Cal Tech has consistently placed as one of the Top 6 Universities since 2012. Cal Tech was bested by Harvard in 2011 by 0.1. While Harvard has always been viewed as the best university in the world, Harvard did not place in Top 6 in 2016. Overall, the scores has been declining for all universities since 2011. This could reflect the changing needs of students.

<b> Number of times a university has been in top 6

```python
top6_times_s = top6_times.reset_index()
top6_times_s['university_name'].value_counts()
```

```python
top6_times_s['university_name'].value_counts().plot(kind = 'bar', figsize = (10,6), rot = 45, title = 'Number of times in Top 6')
plt.ylabel('Count')
plt.xlabel('University Name')
```

Here, we can see that Stanford University  and CalTech has placed in top 6 in every year on record.

<b> Where are the world's top 100 universities from?

```python
top_country = cwur.groupby(['institution', 'country'])['score'].mean().sort_values(ascending = False).head(100)
```

```python
pd_top_c = pd.DataFrame(top_country)
```

```python
pd_top_c_r = pd_top_c.reset_index()
```

```python
pd_top_c_r.head()
```

```python
pd.DataFrame(pd_top_c_r.country.value_counts()).plot(kind = 'bar', legend = False) #try a pie plot here as well
plt.title("Where are the world's top universities from??")
plt.ylabel('Country')
plt.xlabel('Counts')
```

```python
counts_top = pd_top_c_r.country.value_counts()
```

```python
labels = counts_top.index
plt.pie(counts_top, labels = labels, shadow = True, autopct='%1.1f%%')
plt.title("Where are the world's top universities from??")
fig = plt.gcf()
fig.set_size_inches(8,8)
```
